## 1. Get the data provided, check frequency distribution of the features

```python
Wifi_Data = pd.read_excel("C:\\Users\\theco\\Downloads\\Wifi Data.xlsx")
Wifi_Data
```

```python
Wifi_Data['Borough'].value_counts()
```

```python
Wifi_Data1 = Wifi_Data.Borough.value_counts().reset_index(name='Frequency').rename(columns={'index':'Borough'})
Wifi_Data1
```

## a. Plot the providers vs number of all wi-fi hotspots they have in operation? 

```python
merged_df = pd.merge(Wifi_Data, Wifi_Data1)
merged_df.head()
```

```python
sns.displot(data=merged_df, x="Frequency", y="Provider")  

plt.show()
```

```python
sns.histplot(data=merged_df, x="Frequency", y="Provider")

plt.show()
```

```python
sns.histplot(data=merged_df, x="Frequency", hue="Provider", multiple="stack")

plt.show()
```

## b. Plot providers vs number of free wifi hotspots.

```python
sns.histplot(data=merged_df, x="Type", y="Provider")

plt.show()
```

## c. Are there any provider(s) who do not provide free wifi?

Yes! From the plot above, ALTICEUSA, SPECTRUM and Partner do not provide free wifi.

## d. Which Location type (Location_T) are most common and least common?

```python
sns.histplot(data=merged_df, x="Frequency", hue="Location_T", multiple="stack")

plt.show()
```

The outdoor kiosk is the location type that is most common, and this is closely followed by outdoor.
On the other hand, the indoor location type is the least common and this is followed by the library location type.
